```python
"""    

2.Making the most common type of plots using NumPy arrays 
Most of figuring out what kind of plot to use is getting to feel for the data 
seing what kind of plot suit it best.

Matplotlib visualizations are built on NumPy arrays. So in this section we'll build some of the most common types of plots using NumPy arrays.

Line plot - ax.plot() (this is the default plot in matplotlib)
Scatter plot - ax.scatter()
Bar plot - ax.bar()
Histogram plot - ax.hist()
"""

import numpy as np
x = np.linspace(0,10,100)
x[:10]


 #the default plot
fig, ax = plt.subplots()
ax.plot(x,x**2);
```

```python
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));# scatter uses points
```

```python
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
"""
Ploting data directly with pandas


"""
import pandas as pd
car_sales = pd.read_csv("https://raw.githubusercontent.com/mrdbourke/zero-to-mastery-ml/master/data/car-sales.csv")
car_sales["Price"] = car_sales["Price"].replace(r'[\$\,\.]', '', regex=True)
car_sales
```

```python
# remove last two zros column the .00 makes it be considred as a string

car_sales["Price"] = car_sales["Price"].str[:-2]
car_sales
```

```python
#Add a colum date
n = len(car_sales)
print(n)
car_sales["Sale date"] = pd.date_range('1/1/2025',periods=n)
car_sales
```

```python
# Make total sales column (doesn't work, adds as string)
#car_sales["Total Sales"] = car_sales["Price"].cumsum()

# Oops... want them as int's not string
car_sales["Total Sales"] = car_sales["Price"].astype(int).cumsum()
car_sales
```

```python
 #ploting
car_sales.plot(x='Sale date', y= 'Total Sales')
```

```python
#Heart Disease sections
heart_disease = pd.read_csv("https://raw.githubusercontent.com/mrdbourke/zero-to-mastery-ml/master/data/heart-disease.csv")
heart_disease.head(10)
```

```python
heart_disease["age"].plot.hist(bins=  50)
```

```python
#Histogram for each column since nuemric
 #1. arugmen 1 5 wide and 20 inches tall ,subplots =true enas dont smash the chars into one
heart_disease.plot.hist(figsize=( 5,20), subplots= True);
```

```python
 #Perform data analysis on patient over 50
over_50 = heart_disease[ heart_disease["age"] > 50]
over_50
```

```python
 #Creating scatter plot
over_50.plot(
kind="scatter",
x= "age",
y="chol",
c="target",
figsize=( 10,6)

)
```

```python
#Create the sam eplot using subplots
fig, ax = plt.subplots(figsize=(10,6))
# Plot data from the DataFrame to the ax object
over_50.plot(kind="scatter", 
             x="age", 
             y="chol", 
             c="target", 
             ax=ax); # set the target Axes

# Customize the x-axis limits (to be within our target age ranges)
ax.set_xlim([45, 100]);
```
